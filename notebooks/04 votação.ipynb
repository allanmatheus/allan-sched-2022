```python
import itertools

import pandas as pd
import numpy as np
```

```python
def run_election(votes, weights=None):
    if weights is None:
        weights = list(range(len(votes), 0, -1))
    if len(weights) != len(votes):
        raise ValueError('Votes and weights length are different')
    n = sum(weights)
    c = sum((v * w) for (v, w) in zip(votes, weights))
    return c, c > (n / 2)
```

```python
examples = list(itertools.product([0,1], [0,1], [0,1], [0,1]))
df = pd.DataFrame(examples, columns=['3s', '1m', '5m', '10m'])
df.head()
```

```python
elections = [run_election(example) for example in examples]
df['count'] = [election[0] for election in elections]
df['result'] = [election[1] for election in elections]
df
```
